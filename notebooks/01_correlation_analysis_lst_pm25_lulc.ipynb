# LST vs PM2.5 Correlation Analysis for Vidarbha (2005, 2015, 2025)

Analyze the relationship between Land Surface Temperature (LST) and PM2.5 proxy (AOD) in the Vidarbha region across three different years using spatially sampled data exported from Google Earth Engine.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
```

#LULC vs PM2.5 Analysis

```python
#Step 1: Extract LULC Class for Each Point in Your CSV
```

```python
import pandas as pd
import rasterio
import ast
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# 1. Load your sample CSV for 2025 (adjust path as needed)
csv_path = '/content/drive/My Drive/Vidhaarbha_pbl_research/Vidarbha_Sample_LST_AOD_2025.csv'
df = pd.read_csv(csv_path)
```

```python
# === 2. Extract (lon, lat) from .geo column ===
def extract_coords(geo_str):
    geo = ast.literal_eval(geo_str)       # convert text to dict
    return geo['coordinates'][0], geo['coordinates'][1]
```

```python
print(df['.geo'].head(5).tolist())
```

```python
import json

def extract_coords(geo_str):
    geo = json.loads(geo_str)
    return geo['coordinates'][0], geo['coordinates'][1]

df['lon'], df['lat'] = zip(*df['.geo'].apply(extract_coords))
```

```python
# === 3. Open LULC classified map and sample pixel value at each point ===
lulc_path = '/content/drive/My Drive/Vidhaarbha_pbl_research/LULC_RF_classified_2025_new.tif'
with rasterio.open(lulc_path) as src:
    coords = [(row['lon'], row['lat']) for idx, row in df.iterrows()]
    lulc_values = [val[0] for val in src.sample(coords)]
    df['LULC'] = lulc_values
```

```python
# === 4. Drop rows with missing data (if any) ===
aod_col = 'Aerosol_Optical_Depth_Land_Ocean_Mean_Mean'
df_clean = df.dropna(subset=['LULC', aod_col])
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x=df_clean['LULC'], y=df_clean[aod_col])
plt.xlabel('Land Use/Land Cover Class')
plt.ylabel('PM2.5 Proxy (AOD)')
plt.title('PM2.5 Distribution by LULC Class')
plt.xticks(ticks=range(len(labels)), labels=labels, rotation=45)
plt.tight_layout()
plt.show()
```

```python
summary_pm25_2025 = df_clean.groupby('LULC')[aod_col].agg(['count','mean','median','min','max','std'])
summary_pm25_2025.index = labels
print("2025 - PM2.5 Summary:\n", summary_pm25_2025)
```

```python
from scipy.stats import kruskal

labels = ["Forest", "Crop", "Urban", "Water", "Barren", "Grassland"]
aod_col = 'Aerosol_Optical_Depth_Land_Ocean_Mean_Mean'

# Gather PM2.5 values for each LULC class
groups_2025 = [
    df_clean[df_clean['LULC'] == i][aod_col]
    for i in range(len(labels))
]

# Run the Kruskal-Wallis test
stat_2025, pval_2025 = kruskal(*groups_2025)
print(f"2025: Kruskal-Wallis H test statistic = {stat_2025:.3f}, p-value = {pval_2025:.3e}")
```

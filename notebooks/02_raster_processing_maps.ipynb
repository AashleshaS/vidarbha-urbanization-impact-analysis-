```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
```

```python
lst_file_2025 = "/content/drive/My Drive/Vidhaarbha_pbl_research/LST_2025_Vidarbha.tif"
```

```python
with rasterio.open(lst_file_2025) as src:
    data = src.read(1)
    print("Min:", np.nanmin(data))
    print("Max:", np.nanmax(data))
    print("2nd Percentile:", np.nanpercentile(data, 2))
    print("98th Percentile:", np.nanpercentile(data, 98))
    print("Mean:", np.nanmean(data))
    print("Std. Dev.:", np.nanstd(data))
```

```python
with rasterio.open(lst_file_2025) as src:
    data = src.read(1)
    plt.figure(figsize=(10,8))
    im = plt.imshow(data, cmap='inferno', vmin=36.07, vmax=53.46)
    plt.colorbar(im, label='LST (°C)')
    plt.title("Land Surface Temperature - 2025")
    #plt.axis('off')
    plt.show()
```
